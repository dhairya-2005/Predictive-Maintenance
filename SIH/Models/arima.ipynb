```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error
from pmdarima import auto_arima
import matplotlib.pyplot as plt

data = pd.read_csv("../Datasets/arima_data.csv")
data = data.drop(['timestamp'], axis=1)
rul = data['rul'].values

scaler = MinMaxScaler()
rul_scaled = scaler.fit_transform(rul.reshape(-1, 1)).flatten()

# Train-test split (80% train, 20% test sequentially)
train_size = int(len(rul_scaled) * 0.8)
train_data, test_data = rul_scaled[:train_size], rul_scaled[train_size:]

# Fit ARIMA model using auto_arima for parameter selection
print("finding the best ARIMA parameters...")
arima_model = auto_arima(train_data, seasonal=False, trace=True, stepwise=True, suppress_warnings=True)

# Forecast on test data
n_test_steps = len(test_data)
predicted_scaled = arima_model.predict(n_test_steps)

# Inverse transform predictions to original scale
predicted = scaler.inverse_transform(predicted_scaled.reshape(-1, 1)).flatten()
actual = scaler.inverse_transform(test_data.reshape(-1, 1)).flatten()

# Evaluate accuracy
mape = mean_absolute_percentage_error(actual, predicted)
print(f"MAPE (Mean Absolute Percentage Error): {mape * 100:.2f}%")

# Plot actual vs predicted
plt.figure(figsize=(12, 6))
plt.plot(actual, label='Actual RUL', color='blue')
plt.plot(predicted, label='Predicted RUL', color='red')
plt.title('ARIMA Model Predictions')
plt.xlabel('Time')
plt.ylabel('RUL')
plt.legend()
plt.show()
```

```python
N = 5  
future_predictions_scaled = arima_model.predict(N)
future_predictions = scaler.inverse_transform(future_predictions_scaled.reshape(-1, 1)).flatten()
print(future_predictions)

plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(rul), len(rul) + N), future_predictions, label=f'Future {N} Steps', color='green', linestyle='--')
plt.title('Historical and Future RUL Predictions')
plt.xlabel('Time')
plt.ylabel('RUL')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Load and preprocess data
data = pd.read_csv("/content/synthetic_sensor_data_with_rul.csv")
data = data.drop(['timestamp'], axis=1)

# Target variable (RUL)
rul = data['rul'].values

# Scaling RUL
scaler = MinMaxScaler()
rul_scaled = scaler.fit_transform(rul.reshape(-1, 1)).flatten()

# Increase the training data (using 90% of data for training)
train_size = int(len(rul_scaled) * 0.9)
train_data, test_data = rul_scaled[:train_size], rul_scaled[train_size:]

# Feature Engineering - Create lagged features and rolling statistics (moving averages)
def create_lagged_features(data, lag=15):  # Adjusted lag to capture a longer dependency (15 minutes)
    df = pd.DataFrame(data)
    lagged_data = df.shift(lag)
    return lagged_data.dropna()

def create_rolling_features(data, window=30):  # Adjusted rolling window to capture a longer trend (30 minutes)
    df = pd.DataFrame(data)
    df['rolling_mean'] = df[0].rolling(window=window).mean()
    df['rolling_std'] = df[0].rolling(window=window).std()
    return df.dropna()

# Creating lag features (e.g., lag 15) and rolling statistics (e.g., window size 30)
lagged_features = create_lagged_features(rul_scaled, lag=15)
rolling_features = create_rolling_features(rul_scaled, window=30)

# Combine lag and rolling features into a new dataset
features = pd.concat([lagged_features, rolling_features], axis=1).values

# Train-test split (80% train, 20% test sequentially)
train_size = int(len(features) * 0.9)
X_train, X_test = features[:train_size], features[train_size:]
y_train, y_test = rul_scaled[15:train_size+15], rul_scaled[15+train_size:]

# Fit SARIMA model
print("Finding the best SARIMA parameters...")
sarima_model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))  # Example SARIMA order
sarima_model_fitted = sarima_model.fit(disp=False)

# Forecast on test data
n_test_steps = len(test_data)
predicted_scaled = sarima_model_fitted.predict(start=len(train_data), end=len(train_data)+n_test_steps-1)

# Inverse transform predictions to original scale
predicted = scaler.inverse_transform(predicted_scaled.reshape(-1, 1)).flatten()
actual = scaler.inverse_transform(test_data.reshape(-1, 1)).flatten()

# Evaluate accuracy
mape = mean_absolute_percentage_error(actual, predicted)
print(f"SARIMA MAPE: {mape * 100:.2f}%")

# Plot actual vs predicted
plt.figure(figsize=(12, 6))
plt.plot(actual, label='Actual RUL', color='blue')
plt.plot(predicted, label='Predicted RUL', color='red')
plt.title('SARIMA Model Predictions')
plt.xlabel('Time (minutes)')
plt.ylabel('RUL')
plt.legend()
plt.show()
```
