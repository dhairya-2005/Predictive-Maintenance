```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "../Datasets/new_temp.csv" 
df = pd.read_csv(file_path)
df.head()
```

```python
plt.plot(df["RUL"], df["Temperature (C)"], label="Temperature (C)", color="red", alpha=0.7)
plt.xlabel("RUL")
plt.ylabel("Temperature (C)")
plt.title("Temperature vs RUL")
plt.grid(True)
plt.legend()
```

```python
plt.plot(df["RUL"], df["Humidity (%)"], label="Humidity (%)", color="blue", alpha=0.7)
plt.xlabel("RUL")
plt.ylabel("Humidity (%)")
plt.title("Humidity vs RUL")
plt.grid(True)
plt.legend()
```

```python
plt.plot(df["RUL"], df["Current (A)"], label="Current (A)", color="green", alpha=0.7)
plt.xlabel("RUL")
plt.ylabel("Current (A)")
plt.title("Current vs RUL")
plt.grid(True)
plt.legend()
```

```python
plt.plot(df["RUL"], df["Voltage (V)"], label="Voltage (V)", color="orange", alpha=0.7)
plt.xlabel("RUL")
plt.ylabel("Voltage (V)")
plt.title("Voltage vs RUL")
plt.grid(True)
plt.legend()
```

```python
plt.plot(df["RUL"], df["Pressure (hPa)"], label="Pressure (hPa)", color="purple", alpha=0.7)
plt.xlabel("RUL")
plt.ylabel("Pressure (hPa)")
plt.title("Pressure vs RUL")
plt.grid(True)
plt.legend()
```

```python
plt.plot(df["Failure Level (%)"], df["RUL"], label="RUL vs Failure Level (%)", color="blue", alpha=0.7)
plt.xlabel("Failure Level (%)")
plt.ylabel("RUL)")
plt.title("Pressure vs RUL")
plt.grid(True)
plt.legend()
```
